```python
from IPython.display import display, Image, clear_output, HTML 
import time
import random

import requests
from bs4 import BeautifulSoup
import json
import time
```

```python
def add_score(num_1, num_2):
    product = int(num_1+num_2)
    
    style = f"font-size: 50px; ;"
    html_out_1 = HTML(f"""<span style='{style}'>{num_1}</span>""")
    html_out_2 = HTML(f"""<span style='{style}'>{num_2}</span>""")

    print("Please add the numbers shown.")
    time.sleep(1)

    start_time= time.time()

    display(html_out_1)
    time.sleep(1)

    display(html_out_2)
    time.sleep(1)
    clear_output(wait=False)

    
    score=0
    input_ans= int(input("The answer is:"))
    if product == input_ans:
        print("Well done!")
        time.sleep(0.9)
        clear_output(wait=False)
        score+=1
    else:
        print("Sorry, wrong answer!")
        time.sleep(0.9)
        clear_output(wait=False)
        score+=0
    
    end_time= time.time()
    time_taken= end_time- start_time
    
    return score, time_taken
```

```python
numlist_1= [2, 6, 8, 9, 3,
            7, 4, 6, 8, 5]

numlist_2= [9, 5, 7, 6, 7,
            5, 7, 6, 2, 9]

numlist_3= [56, 78, 36, 43, 89, 
            47, 66, 16, 58, 17]

numlist_4= [29, 84, 47, 69, 55,
            19, 54, 97, 78, 66]

numlist_5= [574, 359, 195, 857, 591, 
            682, 378, 274, 368, 756]

numlist_6= [147, 216, 475, 413, 258, 
            319, 734, 559, 458, 129]
```

```python
def run_add_test(numlist_1, numlist_2):
    total_time= 0
    total_score= 0

    n= len(numlist_1)
    
    for i in range (n):
        num_1= numlist_1[i]
        num_2= numlist_2[i]
        score, time_taken= add_score(num_1,num_2)
        clear_output(wait=False)
        total_time+= time_taken
        total_score+= score
    
    print("Total score", total_score)
    time.sleep(1)
    print("Total time taken", total_time)
    time.sleep(1.5)
    clear_output(wait=False)
    return total_score, total_time

```
