```python
import requests
from bs4 import BeautifulSoup
import json
import time

def send_to_google_form(data_dict, form_url):
    ''' Helper function to upload information to a corresponding google form 
        You are not expected to follow the code within this function!
    '''
    form_id = form_url[34:90]
    view_form_url = f'https://docs.google.com/forms/d/e/{form_id}/viewform'
    post_form_url = f'https://docs.google.com/forms/d/e/{form_id}/formResponse'

    page = requests.get(view_form_url)
    content = BeautifulSoup(page.content, "html.parser").find('script', type='text/javascript')
    content = content.text[27:-1]
    result = json.loads(content)[1][1]
    form_dict = {}
    
    loaded_all = True
    for item in result:
        if item[1] not in data_dict:
            print(f"Form item {item[1]} not found. Data not uploaded.")
            loaded_all = False
            return False
        form_dict[f'entry.{item[4][0][0]}'] = data_dict[item[1]]
    
    post_result = requests.post(post_form_url, data=form_dict)
    return post_result.ok
```

```python
def run_ANS_test(image_list, i):
    print("you are on question", i+1, "out of", len(image_list) )
    time.sleep(1.5)
    display(image_list[i][0]) #This displays the question.
    time.sleep(0.75)
    clear_output(wait=False)
    display(Image("0vs0.jpg", width=500))

    result=[] #I created a "result" array to add and store information for each question. Each question will store 3 information.
    start = time.time()
    ans=""
    ans = input("Which side had more dots (r or l)?: ")
    result.append(ans) #The first data in the array will store whatever the answer that the user enters.
    if ans == image_list[i][1]: #This first if statement is if the user enters the correct answer.
        response_time = time.time()-start
        result.append(response_time) #The second data in the array stores how long the user took on each question. 
        if response_time <= 3:
            print("Well done!")
            time.sleep(1)
            print(f"You took {response_time:.2f} seconds")
            result.append(0) #The last data in the array is for counting the number of incorrect answers. Entering the correct answer will just append 0.
        if response_time > 3:
            print("you spent longer than 3 seconds")
            result.append(1) #Taking longer than 3 seconds will be considered as "incorrect" and will add 1 to the result array.
    else: #This is if the user enters an incorrect answer
        print("incorrect")
        time.sleep(1)
        result.append(0) 
        result.append(1) #this is used to count the number of incorrect answers

    time.sleep(1.5)
    clear_output(wait=False)
    return result 
```
