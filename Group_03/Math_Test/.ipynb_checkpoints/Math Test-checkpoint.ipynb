```python
from IPython.display import display, HTML, Image, clear_output
import time
import random
import pandas
import requests
from bs4 import BeautifulSoup
import json

#notes to self, discuss with team members how to increase difficulty and what difficulty is appropriate
#length takes two minutes, more questions or harder questions
#any optimisation

#Has to be global or else can't call from within function
score = 0 

def send_to_google_form(data_dict, form_url):
    ''' Helper function to upload information to a corresponding google form 
        You are not expected to follow the code within this function!
    '''
    form_id = form_url[34:90]
    view_form_url = f'https://docs.google.com/forms/d/e/{form_id}/viewform'
    post_form_url = f'https://docs.google.com/forms/d/e/{form_id}/formResponse'

    page = requests.get(view_form_url)
    content = BeautifulSoup(page.content, "html.parser").find('script', type='text/javascript')
    content = content.text[27:-1]
    result = json.loads(content)[1][1]
    form_dict = {}
    
    loaded_all = True
    for item in result:
        if item[1] not in data_dict:
            print(f"Form item {item[1]} not found. Data not uploaded.")
            loaded_all = False
            return False
        form_dict[f'entry.{item[4][0][0]}'] = data_dict[item[1]]
    
    post_result = requests.post(post_form_url, data=form_dict)
    return post_result.ok

#2 variable display
def df2val(rn1, rn2):
    html_out = HTML(f"""<span style='color: black; font-size: 50px'>{rn1}</span>""")
    display(html_out)
    time.sleep(2)
    clear_output(wait=False)
    #starts timer as soon as second value is displayed
    global start_time
    start_time = time.time()
    html_out = HTML(f"""<span style='color: black; font-size: 50px'>{rn2}</span>""")
    display(html_out)
    time.sleep(2)
    clear_output(wait=False)
    #question function can be changed 

#3 variable display
def df3val(rn1, rn2, rn3):
    html_out = HTML(f"""<span style='color: black; font-size: 50px'>{rn1}</span>""")
    display(html_out)
    time.sleep(2)
    clear_output(wait=False)
    #starts timer as soon as second value is displayed
    global start_time
    start_time = time.time()
    html_out = HTML(f"""<span style='color: black; font-size: 50px'>{rn2}</span>""")
    display(html_out)
    time.sleep(2)
    clear_output(wait=False)
    html_out = HTML(f"""<span style='color: black; font-size: 50px'>{rn3}</span>""")
    display(html_out)
    time.sleep(2)
    clear_output(wait=False)
    #question function can be changed 

#Question and answer
def question(ANSW):
    ans = ""
    ans = int(input("Answer:"))
    if ans == ANSW: 
        global score
        score += 1
        clear_output(wait=False) 
        global end_time
        end_time=time.time()
        html_out = HTML(f"""<span style='color: black; font-size: 50px'>Answer Correct</span>""")
        display(html_out)
    else:
        clear_output(wait=False) 
        html_out = HTML(f"""<span style='color: black; font-size: 50px'>Answer Incorrect</span>""")
        display(html_out)
        end_time=time.time()
    global time_taken
    time_taken = end_time - start_time
    # the time taken to complete question is designated as the time from when the second value is displayed, thus comencing 
    # calculation till the particpant answers the question, it would be pointless to start the timer as soon as values are 
    # displayed as that will add around 2 seconds to every particpants time that cannot be reduced.
    time.sleep(1)
    clear_output(wait=False)

#2 number addition test
def test1():
    #randomises two number and add together to create answer)
    random_num1 = random.randint(1,100)#number here can be varied to increase or increase difficulty
    random_num2 = random.randint(1,100)
    answer = random_num1 + random_num2
    #Adds plus sign infront of interger
    input2 = "{0:+}".format(random_num2)
    df2val(random_num1, input2)
    question(answer)
    return

#2 number multiplication test
def test2():
    random_num1 = random.randint(2, 15)#Increase to 15 for difficulty
    random_num2 = random.randint(2, 15)
    answer = random_num1 * random_num2
    #Adds x infront of interger
    input2 = "x" + str(random_num2)
    df2val(random_num1, input2)
    question(answer)
    return

#3 number addition test
def test3():
    random_num1 = random.randint(1, 100)
    random_num2 = random.randint(1, 100)
    random_num3 = random.randint(1, 100)
    input2 = "{0:+}".format(random_num2)
    input3 = "{0:+}".format(random_num3)
    answer = random_num1 + random_num2 + random_num3
    df3val(random_num1, input2, input3)
    question(answer)
    return

#X x Y + Z
def test4():
    random_num1 = random.randint(2, 20)
    random_num2 = random.randint(2, 15)#same as two number multiplication do not increase max value too much
    random_num3 = random.randint(1, 100)
    answer = random_num1 * random_num2 + random_num3
    input2 = "x" + str(random_num2)
    input3 = "{0:+}".format(random_num3)
    df3val(random_num1, input2, input3)
    question(answer)
    return

#(X + Y) x Z
def test5():
    random_num1 = random.randint(1, 50)
    random_num2 = random.randint(1, 50)
    random_num3 = random.randint(2, 15)#setting max to 15 to increase difficult can be viable, going above would be pointless
    input2 = "{0:+}".format(random_num2)
    input3 = "x" + str(random_num3)
    answer = (random_num1 + random_num2) * random_num3
    df3val(random_num1, input2, input3)
    question(answer)
    return

# 3 number multiplication
def test6():
    random_num1 = random.randint(2, 10)
    random_num2 = random.randint(2, 10)
    random_num3 = random.randint(2, 10)
    input2 = "x" + str(random_num2)
    input3 = "x" + str(random_num3)
    answer = (random_num1 * random_num2 * random_num3)
    df3val(random_num1, input2, input3)
    question(answer)
    return

def runtest():
    total_time = 0
    html_out = HTML(f"""<span style='color: black; font-size: 50px'>Welcome to the math ability test</span>""")
    display(html_out)
    time.sleep(2)
    clear_output(wait=False)
 
    #name input
    html_out = HTML(f"""<span style='color: black; font-size: 50px'>What is your name?</span>""")
    display(html_out)
    username = input(">")
    time.sleep(2)
    clear_output(wait=False)
    
    #sex input
    html_out = HTML(f"""<span style='color: black; font-size: 50px'>What is your biological sex? Input Male, Female or Other</span>""")
    display(html_out)
    usersex = input(">")
    time.sleep(2)
    clear_output(wait=False)
    
    #arbritary stuff
    html_out = HTML(f"""<span style='color: black; font-size: 50px'>Your test will include 6 parts with 5 questions each</span>""")
    display(html_out)
    time.sleep(2)
    clear_output(wait=False)
    html_out = HTML(f"""<span style='color: black; font-size: 50px'>Your score and time will be recorded</span>""")
    display(html_out)
    time.sleep(2)
    clear_output(wait=False)
    html_out = HTML(f"""<span style='color: black; font-size: 50px'>Part 1 starting now</span>""")
    display(html_out)
    time.sleep(2)
    clear_output(wait=False)

    #start of test
    for i in range (5):
        test1()
        total_time += time_taken #updates total_time from time_taken
        time.sleep(2)
    html_out = HTML(f"""<span style='color: black; font-size: 50px'>Part 2 starting now</span>""")
    display(html_out)
    time.sleep(2)
    clear_output(wait=False)
    for i in range (5):
        test2()
        total_time += time_taken
        time.sleep(2)
    html_out = HTML(f"""<span style='color: black; font-size: 50px'>Part 3 starting now</span>""")
    display(html_out)
    time.sleep(2)
    clear_output(wait=False)
    for i in range (5):
        test3()
        total_time += time_taken
        time.sleep(2)
    html_out = HTML(f"""<span style='color: black; font-size: 50px'>Part 4 starting now</span>""")
    display(html_out)
    time.sleep(2)
    clear_output(wait=False)
    for i in range (5):
        test4()
        total_time += time_taken
        time.sleep(2)
    html_out = HTML(f"""<span style='color: black; font-size: 50px'>Part 5 starting now</span>""")
    display(html_out)
    time.sleep(2)
    clear_output(wait=False)
    for i in range (5):
        test5()
        total_time += time_taken
        time.sleep(2)
    html_out = HTML(f"""<span style='color: black; font-size: 50px'>Part 6 starting now</span>""")
    display(html_out)
    time.sleep(2)
    clear_output(wait=False)
    for i in range (5):
        test6()
        total_time += time_taken
        time.sleep(2)

    #result display
    ave_time = total_time/24#alter the denominater depending on how much questions there are
    print("Your score is: ", score)
    print("You took:  ", total_time, "seconds to complete the test")
    print("On average you took:", ave_time, "to complete each question")

    #send data to form"
    data_dict = {
    'name': username,
    'sex': usersex,
    'playerscore': score,
    'averagetime': ave_time
    }
    form_url = 'https://docs.google.com/forms/d/e/1FAIpQLSeoPgA0CDrG8R55Z6bSmt5j3s0CR9kobdhmRp5mALF2mLdUsw/viewform?usp=sf_link'
    
    #data privacy
    print("Please read:")
    print("")
    print("we wish to record your response data")
    print("to an anonymised public data repository. ")
    print("Your data will be used for educational teaching purposes")
    print("practising data analysis and visualisation.")
    print("")
    print("Please type   yes   in the box below if you consent to the upload.")
    result = input("> ")
    if result == "yes":
        print("Thanks - your data will be uploaded.")
        send_to_google_form(data_dict, form_url)
    else:
        print("No problem we hope you enjoyed the test.")

runtest()
```
