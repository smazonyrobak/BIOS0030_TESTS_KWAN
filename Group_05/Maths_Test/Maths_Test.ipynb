```python
import random 
import time
from IPython.display import clear_output
random.seed(10)

#INTRODUCING THE TEST CODE 

#user id 
id_instructions = """

Hello, before we begin please enter your anonymised ID

To generate an anonymous 4-letter unique user identifier please enter:

- two letters based on the initials (first and last name) of a childhood friend

- two letters based on the initials (first and last name) of a favourite actor / actress

e.g. if your friend was called Charlie Brown and film star was Tom Cruise

then your unique identifer would be CBTC

"""

print(id_instructions)
user_id = input("> ")
print("User entered id:", user_id)
clear_output(wait=False)
time.sleep(0.1)


#male/female info

sex_info = """

Do you identify as male or female? 

Please enter:
- M for Male 
- F for Female 
- P for prefer not say 

""" 

print(sex_info)
user_sex = input ()
clear_output(wait=False)
time.sleep(0.1)


#age info

age_info = """
How old are you? 

""" 

print(age_info)
user_age = input ()
clear_output(wait=False)
time.sleep(0.1)


#course info

course_info = """
What course are you on? 

""" 

print(course_info)
user_course = input ()
clear_output(wait=False)
time.sleep(0.1)



#tiredness info

tired_info = """

How tired are you feeling today?

Enter your answer as a number 1-5

 1 - actually falling asleep right now

 2 - significantly tired

 3 - a little tired

 4 - not tired

 5 - totally wired and almost too awake

"""


print(tired_info)
user_tired = input ()
clear_output(wait=False)
time.sleep(0.1)

#data consent 
data_consent_info = "DATA CONSENT INFORMATION:"

print("Please read:")
print("We wish to record your response data to an anonymised public data repository.")
print("Your data will be used for educational teaching purposes, practising data analysis and visualisation.")
print("Please type yes in the box below if you consent to the upload.")

print(data_consent_info)
result = input("> ")

if result == "yes" or "Yes" or "YES":
    print("Thanks for your participation.") 
    print("Please contact philip.lewis@ucl.ac.uk")
    print("If you have any questions or concerns")
    print("regarding the stored results.")
    print ("now let's continue...")

else:
    raise(Exception("User did not consent to continue test."))
    
clear_output(wait=False)
time.sleep(0.1)


#explaining the test     
welcome_message = """

Welcome to the maths ability test !

you will be presented with maths questions and go through 3 levels that increase in diffculty 

values will only be displayed for 2 seconds and disappear before the next part appears

so make sure you make a mental note of them!

your final score is a combination of your accuracy and how long it takes you to answer the questions

now lets start the test

""" 

print (welcome_message)
input("press enter to continue")



#ACTUAL TEST CODE 

s_operators = ["+", "-"]
c_operators = [ "/", "*"]
a_operators = ["+", "-", "*", "/"]
score = 0

time_per_question_level_1 = []
time_per_question_level_2 = []
time_per_question_level_3 = []

#function to calculate and print average time per question
def print_avg_time_per_question(time_list, level):
    avg_time = sum(time_list) / len(time_list) if len(time_list) > 0 else 0
    print(f"Average time per question in Level {level}: {avg_time:.2f} seconds")

#function to display question parts
def display_question(question_parts):
    for part in question_parts:
        print(part)
        time.sleep(2)
        clear_output(wait=True)



start_time = time.time()


# LEVEL 1
print('Now Level 1 will commence!')
time.sleep(2)
clear_output(wait=False)

for i in range(1, 6):
    start_time_question = time.time()  
    num1 = random.randint(1, 10)
    num2 = random.randint(1, 10)
    op = random.choice(s_operators)
    problem = f"Q{i}. {num1} {op} {num2}?"

    if op == "+":
        answer = num1 + num2
    elif op == "-":
        answer = num1 - num2

    question_parts = [f"Q{i}. What is {num1} ", f"{op} {num2}?"]
    display_question(question_parts)

    user_answer = input()

    if float(user_answer) == answer:
        print("Correct!")
        score += 1
    else:
        print("Incorrect.")

    time.sleep(2)
    clear_output(wait=True)
    end_time_question = time.time() 
    time_taken = end_time_question - start_time_question 
    time_per_question_level_1.append(time_taken) 

    
# LEVEL 2
print('Now Level 2 will commence')
time.sleep(2)
clear_output(wait=False)

for a in range(1, 6):
    start_time_question = time.time()  
    num1 = random.randint(1, 50)  
    num2 = random.randint(1, 10)  

    op = random.choice(c_operators)
    problem = f"Q{a}. {num1} {op} {num2}?"
    
    if op == "*":
        answer = num1 * num2
    elif op == "/":
        while num2 >= num1 or num1 % num2 != 0: #to make sure divison can still be done mentally, ensures num1 > num2 and the division results in no remainder 
            num1 = random.randint(1, 50)
            num2 = random.randint(1, 10)
        answer = num1 // num2  

    question_parts = [f"Q{a}. What is {num1} ", f"{op} {num2}?"]
    display_question(question_parts)

    user_answer = input()

    if float(user_answer) == answer:
        print("Correct!")
        score += 2
    else:
        print("Incorrect.")

    time.sleep(2)
    clear_output(wait=True)
    end_time_question = time.time()  
    time_taken = end_time_question - start_time_question  
    time_per_question_level_2.append(time_taken)  
    

# LEVEL 3  
print('Now Level 3 will commence, there will be 3 variables in these questions ie 3 * 4 - 2')
time.sleep(2)
clear_output(wait=False)

for a in range(1, 6):
    start_time_question = time.time() 
    num1 = random.randint(1, 50)  
    num2 = random.randint(1, 10)  
    num3 = random.randint(1, 10)  

    op1 = random.choice(a_operators)
    op2 = random.choice(a_operators)
    
   
    problem = f"Q{a}. {num1} {op1} {num2} {op2} {num3}?"
    
  
    if op1 == "*":
        intermediate = num1 * num2
    elif op1 == "/":
        while num2 >= num1 or num1 % num2 != 0:
            num1 = random.randint(1, 50)
            num2 = random.randint(1, 10)
        intermediate = num1 // num2
    elif op1 == "+":
        intermediate = num1 + num2
    else:
        intermediate = num1 - num2

    if op2 == "*":
        answer = intermediate * num3
    elif op2 == "/":
        while num3 >= intermediate or intermediate % num3 != 0:
            num3 = random.randint(1, 10)
        answer = intermediate // num3
    elif op2 == "+":
        answer = intermediate + num3
    else:
        answer = intermediate - num3
    
    question_parts = [f"Q{a}. What is {num1}",f" {op1} {num2}", f" {op2} {num3}?"]
    display_question(question_parts)

    user_answer = input()

    if float(user_answer) == answer:
        print("Correct!")
        score += 3
    else:
        print("Incorrect.")

    time.sleep(2)
    clear_output(wait=True)
    end_time_question = time.time() 
    time_taken = end_time_question - start_time_question 
    time_per_question_level_3.append(time_taken) 



end_time = time.time()
total_time = end_time - start_time


#OUTPUT CODE 

#time multiplier rule for score
if total_time <= 90: 
    multiplier = 2 # multiply accuracy score by 2 if test finishes in under 90 seconds 
elif total_time <= 120:  
    multiplier = 1.5 # multiply accuracy score by 1.5 if test finishes in under 120 seconds 
elif total_time <= 150: 
    multiplier = 1.25 # multiply accuracy score by 1.25 if test finishes in under 150 seconds 
else: 
    multiplier = 1 #if time taken is longer than 150 seconds, there will be no time bonus 

score *= multiplier


# calculate overall average time per question
all_time_per_question = time_per_question_level_1 + time_per_question_level_2 + time_per_question_level_3
total_time = sum(all_time_per_question)
total_questions = len(all_time_per_question)

overall_avg_time_per_question = total_time / total_questions if total_questions > 0 else 0


# what user is told 
print(f"{user_id} your score was {score}!")
print(f"you took {total_time:.2f} seconds overall")
print(f"your average time per question across all levels was {overall_avg_time_per_question:.2f} seconds")


#GOOGLE FORM CODE 

import requests
from bs4 import BeautifulSoup
import json

def send_to_google_form(data_dict, form_url):
    ''' Helper function to upload information to a corresponding google form 
        You are not expected to follow the code within this function!
    '''
    form_id = form_url[34:90]
    view_form_url = f'https://docs.google.com/forms/d/e/{form_id}/viewform'
    post_form_url = f'https://docs.google.com/forms/d/e/{form_id}/formResponse'

    page = requests.get(view_form_url)
    content = BeautifulSoup(page.content, "html.parser").find('script', type='text/javascript')
    content = content.text[27:-1]
    result = json.loads(content)[1][1]
    form_dict = {}
    
    loaded_all = True
    for item in result:
        if item[1] not in data_dict:
            print(f"Form item {item[1]} not found. Data not uploaded.")
            loaded_all = False
            return False
        form_dict[f'entry.{item[4][0][0]}'] = data_dict[item[1]]
    
    post_result = requests.post(post_form_url, data=form_dict)
    return post_result.ok





data_dict = {
    'user id': user_id,
    'user course': user_course,
    'user sex' : user_sex,
    'user age': user_age,
    'user tiredness': user_tired,
    'total score': score,
    'avg time': overall_avg_time_per_question
    
} 


form_url = 'https://docs.google.com/forms/d/e/1FAIpQLSdm527RHRRRTIJxvJ2mkMswkTXknlzjBLR7C2IHgrwU11Mb0g/viewform?usp=sf_link'
send_to_google_form(data_dict, form_url)







```
